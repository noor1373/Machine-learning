Programmet rekommenderar fem filmer som liknar en film du väljer. Om du t.ex. skriver in "Toy Story" så hittar programmet fem andra filmer som påminner om den baserat på genre och taggar. Det använder två filer: movies.csv med filmtitlar och genrer, samt tags.csv med ord som användare skrivit om filmer.

```python
from recommender import initialize
initialize()
```

Genres i datasetet separeras med pipe-tecken, t.ex. Comedy|Drama|Romance.
Pipe-tecknen tas bort så att det blir vanliga ord separerade med mellanslag.
Alla taggar en användare skrivit om en film slås sedan ihop till ett enda textfält per film som kombineras med genrerna.

```python
from recommender import load_data
movies = load_data()
movies[['title', 'genres', 'tag_text', 'combined_text']].head(10)
```

TF-IDF omvandlar texten till siffror. Vanliga ord som "Drama" får lågt värde eftersom de förekommer i många filmer, medan ovanliga ord som "surrealism" får högt värde och bidrar mer till jämförelsen. KNN med cosine similarity används sedan för att hitta de filmer vars TF-IDF-vektor ligger närmast varandra.

```python
from recommender import build_model
tfidf_matrix, model = build_model(movies)
print(f'Matrisform: {tfidf_matrix.shape}, mått: {model.metric}')
```

Funktionen get_recommendations() tar emot en filmtitel, söker upp filmen i datasetet och returnerar de fem mes liknande filmerna sorterade efter likhetspoäng.

```python
from recommender import get_recommendations
display(get_recommendations('Toy Story (1995)', n=5))
```

Programmer kombinerar genrer och användartaggar, omvandlar dem till siffror med TF-IDF och använder KNN för att hitta liknande filmer. Det är en enkel men effektiv metod som ger bättre resultat än att bara jämföra genrer.
